## Week 1: Introduction to Python

```python
# Task: Calculator and Temperature Processor
def calculator():
    num1 = float(input("Enter first number: "))
    op = input("Enter operator (+, -, *, /): ")
    num2 = float(input("Enter second number: "))
    if op == '+': print(f"Result: {num1 + num2}")
    elif op == '-': print(f"Result: {num1 - num2}")
    elif op == '*': print(f"Result: {num1 * num2}")
    elif op == '/': print(f"Result: {num1 / num2}" if num2 != 0 else "Error")

def avg_temp():
    temps = [25, 28, 30, 22, 24, 27, 29]
    print(f"Average Weekly Temp: {sum(temps)/len(temps):.2f}°C")

if __name__ == "__main__":
    avg_temp()
    calculator()
```

## Week 2: Data Structures & Functions

```python
# --- Week 2: Data Structures & Cleaning ---

# 1. Hands-On: Working with Data Structures
def data_transformations():
    numbers = [1, 2, 3, 4, 5, 6]
    # Sum of squares using list comprehension
    sum_squares = sum([x**2 for x in numbers])
    # Filtering even numbers
    filtered_list = [x for x in numbers if x % 2 == 0]

    print("Original List:", numbers)
    print("Sum of Squares:", sum_squares)
    print("Filtered (Evens):", filtered_list)

# 2. Client Project: Data Cleaning Script
def clean_client_data():
    # Sample dataset with duplicates
    raw_data = [
        {"id": 101, "name": "Aman", "city": "Trichy"},
        {"id": 102, "name": "Rock", "city": "Chennai"},
        {"id": 101, "name": "Aman", "city": "Trichy"}, # Duplicate
        {"id": 103, "name": "Rahul", "city": "Madurai"}
    ]

    # Remove duplicates
    unique_data = []
    seen_ids = set()

    for person in raw_data:
        if person["id"] not in seen_ids:
            unique_data.append(person)
            seen_ids.add(person["id"])

    print("\n--- Data Cleaning Results ---")
    print(f"Original Count: {len(raw_data)}")
    print(f"Cleaned Count: {len(unique_data)}")
    print("Cleaned Data:", unique_data)

# Run the Week 2 tasks
if __name__ == "__main__":
    data_transformations()
    clean_client_data()
```

## Week 3: NumPy and Pandas

```python
import pandas as pd
import numpy as np

# Week 3 Task: Load and process data
data = {
    'Product': ['Laptop', 'Mouse', 'Monitor', 'Laptop', 'Mouse'],
    'Sales': [1200, 25, 300, 1200, np.nan], # Missing value
    'Quantity': [5, 10, 7, 5, 2]
}
df = pd.DataFrame(data)

# Handle missing values & aggregate
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
summary = df.groupby('Product')['Sales'].sum().reset_index()

print("--- Week 3 Summary ---")
print(summary)
```

## Week 4: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visual style
sns.set_theme(style="whitegrid")

# Create the plot
plt.figure(figsize=(8, 5))
plot = sns.barplot(x='Product', y='Sales', data=summary, palette='viridis')

# Add labels and title
plt.title('Total Sales Analysis by Product (Month 1)', fontsize=15)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)

# Show the plot
plt.show()
```
